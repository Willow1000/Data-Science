## Pandas Series

```python
order = [1,2,3,4,5,6]
array = [1,2,3,4,5,6]
s = pd.Series(array,index=order)
s
# pd.Series can be used with dictionaries
```

```python
# Appending Series
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,])
s3 = s1.append(s2)

# changing the index of series
index = ['A','B','C','D','E','F','G','H','I']
s3.index = index

# Adding Series
s1.add(s2)

# Getting median
s3.median()
# Getting maximum value
s3.max()
# Getting minimum value
s3.min()
# Getting mean
s3.mean()
# Getting total
s3.sum()
# Getting standard deviation
s3.std()
# Division
s2.div(s1)
# Multiplication
s2 *s1
```
