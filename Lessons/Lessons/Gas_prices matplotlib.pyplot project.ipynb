```python
import pandas as pd
import numpy as np
```

```python
gas_prices = pd.read_csv('gas_prices.csv')
gas_lysis=gas_prices.dropna()
```

## Bar Graph 

```python
import matplotlib.pyplot as mpt
mpt.style.use('ggplot')
years = gas_lysis.groupby('Australia').sum()['Year']
values=[value for value, df in gas_lysis.groupby('Australia')]
mpt.bar(years,values,color = 'yellow')
mpt.xticks(years, rotation = 'vertical',size=9,color = 'red')
mpt.title("Australia's Gas Prices 1991-2008",fontdict={'fontsize':14,'fontweight':'bold'})
mpt.xlabel('Years',fontdict = {'fontname':'Calibri','fontsize':9,'color':'blue'})
mpt.ylabel('Price in USD per Gallon',fontdict = {'fontname':'Calibri','fontsize':9,'color':'blue',})
mpt.show()
```

## Comparative Line graph

##### Tideous and time consuming

```python
gas=gas_lysis
```

```python
mpt.figure(figsize = (5,4),dpi=120)
mpt.plot(gas.Year,gas.Australia,'b.-',label= 'Australia')
mpt.xticks(gas.Year,rotation='vertical',color='red')
mpt.plot(gas.Year,gas.USA,'r.-',label= 'USA')
mpt.plot(gas.Year,gas['South Korea'],'g.-',label='South Korea')
mpt.xlabel('Year',fontdict = {'fontname':'Calibri','fontsize':9,'color':'blue'})
mpt.ylabel('Price per gallon in USD',fontdict = {'fontname':'Calibri','fontsize':9,'color':'blue'})
mpt.title('Gas prices from 1990 to 2008',fontdict = {'fontsize':13,'fontweight':'bold'})
mpt.legend()
mpt.show()
```
