# 3D

```python
# Contour image(map)
import matplotlib.cm as cm
delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X, Y = np.meshgrid(x,y)
Z1=np.exp(-X**2 - Y**2)
Z2=np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig,ax=plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Contour map'.upper())
```

### 3D surface image

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))
ax=fig.add_subplot(1,1,1,projection='3d')
ax.plot_surface(X,Y,Z, rstride=4, cstride=4, linewidth=0)
```

```python
from bokeh.transform import cumsum 
import math
import re
import os
import pandas as pd
import pandas_bokeh
import glob
from bokeh.io import show,output_notebook
from bokeh.plotting import figure
pandas_bokeh.output_notebook()
pd.set_option('plotting.backend','pandas_bokeh')
```

```python
a_us=pd.read_csv('amongus.csv')
# a_us.dropna(inplace=True)
a_us.describe(include='O')
a_us['Minute'] = a_us['Game Length'].apply(lambda x: x.split(' ')[0])
a_us['Minute'] = a_us.Minute.replace('m','',regex=True)
# a_us['Minute']=a_us.Minute.astype('int64')
# a_us['Task Completed'] = a_us['Task Completed'].astype('int64')
# a_us['Sabotages Fixed'] = a_us['Sabotages Fixed'].astype('float64')
a_us['Murdered'].replace(['Yes','No','-'],['Was Killed','Not killed','Missing'],inplace=True)
a_us.replace('-','missing',inplace=True)
```

```python
# PIE CHART
a_us=pd.read_csv('amongus.csv')
a_us_team = a_us.Team.value_counts()
a_us_team.plot_bokeh(kind='pie',title='Crewmates vs Imposters')
```

```python
# Donut plot
a_us_murdered = a_us.Murdered.value_counts().reset_index().rename(columns={'index':'Murdered','Murdered': 'Value'})
a_us_murdered['Angle'] = a_us_murdered['Value']/a_us_murdered['Value'].sum() * 2*math.pi
a_us_murdered['Color'] = ['#fcba03', '#24fc03', '#03e8fc']
a_us_murdered


fig = figure(plot_height=350,toolbar_location=None,
tools = "hover" ,tooltips="@Murdered: @Value" ,x_range=(-0.5,.5))
fig.annular_wedge(x=0,y=1,inner_radius=.15,
                  outer_radius=.25, direction='anticlock',start_angle=cumsum('Angle',include_zero=True),
                end_angle= cumsum('Angle'),line_color='white',fill_color='Color',legend='Murdered',source=a_us_murdered)

fig.axis.axis_label = None
fig.axis.visible=False
fig.grid.grid_line_color=None
show(fig)
```
