```python
legendary=fifa_dat.loc[fifa_dat.BallControl >= 85].count()[0]
intermediates = fifa_dat.loc[(fifa_dat.BallControl>=75) & (fifa_dat.BallControl<85)].count()[0]
Amateurs = fifa_dat.loc[(fifa_dat.BallControl>=60) & (fifa_dat.BallControl<75)].count()[0]
novices = fifa_dat.loc[(fifa_dat.BallControl >=50) & (fifa_dat.BallControl<60)].count()[0]
Begginers = fifa_dat.loc[fifa_dat.BallControl <50].count()[0]
categories = [legendary,intermediates,Amateurs,novices,Begginers]
labels = ['legendary','intermediates','Amateurs','novices','Begginers']
explode = [.5,.0,.0,.0,.0]
# mpt.pie(categories,labels=labels,autopct='%.2f %%',explode=explode)
fig,axes=plt.subplots(1,2,figsize=(16,5))
axes[0].pie(categories,labels=labels,autopct='%.2f %%',explode=explode)
axes[1].bar(labels,categories)
mpt.show()
mpt.style.use('ggplot')
```

```python
sales=pd.read_csv('all_data')
sales.dropna(inplace=True)
sales[sales.isna().any(True)]
sales['City'] = sales['Purchase Address'].apply(lambda x:x.split(',')[1] + ' ' + str(x.split(',')[2]).split(' ')[1])
sales['Quantity Ordered'].sum()
```

```python
quantity=sales.groupby('Product').sum()['Quantity Ordered']
product = [pr for pr, df in sales.groupby('Product')]
mpt.bar(product,quantity)
mpt.xticks(rotation='vertical')
mpt.show()
```

```python
vgs=pd.read_csv('vgsales.csv')
vgs=vgs.sort_values('Global_Sales',ascending=False)
vgs.Publisher.drop_duplicates()
sales=vgs.loc[(vgs.Publisher=='Microsoft Game Studios') & (vgs.Global_Sales>=5)].groupby('Genre').sum()['Global_Sales']
genre = [Genre for Genre,df in vgs.loc[(vgs.Publisher=='Microsoft Game Studios') & (vgs.Global_Sales>=5)].groupby('Genre')]
explode= [.2,.3,.9,.2,.2,.2]
mpt.pie(sales,autopct='%.2f %%',labels=genre,pctdistance=.8,explode=explode)
mpt.show()
```
