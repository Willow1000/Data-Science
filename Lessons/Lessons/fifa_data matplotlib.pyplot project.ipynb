```python
import pandas as pd
import matplotlib.pyplot as mpt
import numpy as np
```

```python
fd = pd.read_csv('fifa_data.csv')
```

### Histogram

```python
mpt.style.use('ggplot')
mpt.figure(figsize=(9,6),dpi=70)
bins = [40,50,60,70,80,90,100]
mpt.hist(fd.Overall,bins=bins,color='#03fcb1')
mpt.xticks(bins)
mpt.xlabel('Overall Rating',fontdict = {'fontname':'Calibri','fontsize':9,'color':'blue'})
mpt.ylabel('Number of players',fontdict = {'fontname':'Calibri','fontsize':9,'color':'blue'})
mpt.title('Distribution of palyers according to overall ratings in fifa 18',fontdict = {'fontsize':13,'fontweight':'bold'})

mpt.show()
```

## Pie chart

```python
left = fd.loc[fd['Preferred Foot']=='Left'].count()[0]
right = fd.loc[fd['Preferred Foot']=='Right'].count()[0]
colors = ['#03fc0b','#2803fc']
labels = ['left', 'right']
mpt.pie([left,right],colors=colors,autopct="%.2f %%",labels=labels)
mpt.title('Foot preferrence of players in fifa 18')
mpt.legend()
mpt.show()
```

```python

mpt.style.use('ggplot')
fd.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fd.Weight]
light = fd.loc[fd.Weight < 125].count()[0]
medium_light = fd.loc[(fd.Weight >= 125) & (fd.Weight <= 150)].count()[0]
medium = fd.loc[(fd.Weight >= 150) & (fd.Weight <= 175)].count()[0]
medium_heavy = fd.loc[(fd.Weight >= 175) & (fd.Weight <= 210)].count()[0]
heavy=fd.loc[fd.Weight >= 210].count()[0]
masses = [light,medium_light,medium,medium_heavy,heavy]
labels=['Under 125','125-150','150-175','175-210','above 210']
explode=[.4,.2,0,0,0.4]
mpt.pie(masses,autopct='%.2f %%',labels=labels,pctdistance=0.8,explode=explode)
mpt.title('Distribution of players according to their weights in lbs')


mpt.show()
```
