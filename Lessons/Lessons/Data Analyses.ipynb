## OBJECTIVE: Find city with highest sales

##### Adding a city column

```python
def get_city(address):
    return address.split(',')[1] + (address.split(',')[2]).split(" ")[1]
all_data['city'] = all_data['Purchase Address'].apply(lambda x: get_city(x))
# OR
# all_data['city'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data
```

###### Answer

```python

outcome = all_data.groupby('city').sum()
# cities = all_data['city'].unique() - extracts items from city column while avoiding duplicates 
cities  =[city for city, df in all_data.groupby('city')]
mpt.bar(cities, outcome['sales'],color='blue')
mpt.xticks(cities, rotation = 'vertical', size = 9, color = 'red')
mpt.xlabel('city',fontdict = {'fontsize':10,'fontname':'Calibri','color':'blue'})
mpt.ylabel('sales',fontdict = {'fontsize':10,'fontname':'Calibri','color':'blue'})
mpt.title('sales/city', fontdict = {'fontsize':14})
mpt.figure(figsize = (4,3), dpi = 60)
mpt.show()
```

## OBJECTIVE: Find products that are mostly sold together

#### Grouping up products with same order ID and cleaning off duplicates and placing them in a grouped column

```python
df = all_data.loc[all_data['Order ID'].duplicated(keep=False)]
df['grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID','grouped']].drop_duplicates()
df
```

#####  Answer

```python
from itertools import combinations 
from collections import Counter 
count = Counter()
for row in df['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
for key, value in count.most_common(10):
    print(key,value)
```

```python
mpt.figure(figsize=(10,8),dpi=50)
mpt.style.use('ggplot')
fd.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fd.Weight]
light = fd.loc[fd.Weight < 125].count()[0]
medium_light = fd.loc[(fd.Weight >= 125) & (fd.Weight <= 150)].count()[0]
medium = fd.loc[(fd.Weight >= 150) & (fd.Weight <= 175)].count()[0]
medium_heavy = fd.loc[(fd.Weight >= 175) & (fd.Weight <= 210)].count()[0]
heavy=fd.loc[fd.Weight >= 210].count()[0]
masses = [light,medium_light,medium,medium_heavy,heavy]
labels=['Under 125','125-150','150-175','175-210','above 210']
mpt.pie(masses,autopct='%.2f %%',labels=labels)

mpt.legend()
mpt.show()
```
